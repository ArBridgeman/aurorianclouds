```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
```

```python
unique_cats = (recipe_df.categories.explode().sort_values().unique().tolist())
unique_tags = (recipe_df.tags.explode().unique().tolist())
```

```python
tag_freq = recipe_df.tags.value_counts()
```

```python
tag_leads = defaultdict(int)
for tag in unique_tags:
    if isinstance(tag, str):
        lead = tag.split("/")[0]
        tag_leads[lead] += 1
    
print(sorted(tag_leads.items(), key=lambda x: x[1], reverse=True))
```

```python
sel_leads = ["cuisine", "cuisine-region", "protein", "breakfast", "joint-snack", "dessert"]
```

```python
recipe_df["ingredients_html"] = recipe_df.ingredients.str.replace("\n", "<br>")
```

```python
dropdown_cat = widgets.SelectMultiple(options = ["All"] + unique_cats, description='Categories:',
                                      disabled=False, value=["All"], continuous_update=False)

multi_dropdowns_sel = []
multi_dropdowns_desel = []
for tag_group in sel_leads:
    multi_dropdowns_sel.append(
        widgets.SelectMultiple(
        options=["All"] + [tag for tag in unique_tags if f"{tag_group}/" in str(tag)],
        value=["All"],
        description=tag_group,
        disabled=False,
        continuous_update=False
        )
    )

    multi_dropdowns_desel.append(
        widgets.SelectMultiple(
            options=["None"] + [tag for tag in unique_tags if f"{tag_group}/" in str(tag)],
            value=["None"],
            description=tag_group,
            disabled=False,
            continuous_update=False
        )
    )

time_slider = widgets.IntSlider(
    value=45, # Initial value
    min=1,    # Minimum value
    max=300,   # Maximum value
    step=1,   # Step size
    description='Total time (min):',
    continuous_update=True
)

# text_tag1 = widgets.Text(value='', description='Tags 1 (or):', continuous_update=False, disabled=False)
# text_tag2 = widgets.Text(value='', description='Tags 2 (or):', continuous_update=False, disabled=False)
# text_tag3 = widgets.Text(value='', description='Tags 3 (or):', continuous_update=False, disabled=False)

col_sel = ["title", "rating", "time_total", "ingredients", "categories", "tags"]
sort_cols = ["rating"]
top_show = 5

# Create output widget
output = widgets.Output()

def dropdown_eventhandler(change):
    with output:
        clear_output()
        mask = np.ones(len(recipe_df), dtype=bool)
        
        mask &= ((recipe_df['time_total'] <= pd.Timedelta(f"{time_slider.value} min")) | recipe_df['time_total'].isna())

        if dropdown_cat.value != ('All',):
            sub_mask = np.zeros(len(mask), dtype=bool)
            all_vals = dropdown_cat.value
            for val in all_vals:
                if val != 'All':
                    sub_mask |= recipe_df['categories'].astype(str).str.contains(val.strip())
            mask &= sub_mask
        
        for dropdown in multi_dropdowns_sel:
            if dropdown.value != ('All',):
                sub_mask = np.zeros(len(mask), dtype=bool)
                all_vals = dropdown.value
                for val in all_vals:
                    if val != 'All':
                        sub_mask |= recipe_df['tags'].astype(str).str.contains(val.strip())
                mask &= sub_mask

        for dropdown in multi_dropdowns_desel:
            if dropdown.value != ('None',):
                sub_mask = np.ones(len(mask), dtype=bool)
                all_vals = dropdown.value
                for val in all_vals:
                    if val != 'None':
                        sub_mask &= ~recipe_df['tags'].astype(str).str.contains(val.strip())
                mask &= sub_mask
                
        display(
            HTML(
                recipe_df[mask][col_sel].sort_values(by=sort_cols, ascending=False).head(top_show).T.to_html(index=False).replace("\\n", "<br>")
             )
        )
    
dropdown_cat.observe(dropdown_eventhandler, names='value')
# Bind function to text box value change
time_slider.observe(dropdown_eventhandler, names='value')

for multi_drop in multi_dropdowns_sel:
    multi_drop.observe(dropdown_eventhandler, names='value')

for multi_drop in multi_dropdowns_desel:
    multi_drop.observe(dropdown_eventhandler, names='value')

display(dropdown_cat)

sel_lab = widgets.Label(value='Include')
desel_lab = widgets.Label(value='Exclude')

hbox_select = widgets.HBox([sel_lab] + multi_dropdowns_sel)
hbox_deselect = widgets.HBox([desel_lab] + multi_dropdowns_desel)

display(hbox_select)
display(hbox_deselect)

display(time_slider)

display(output)
```
